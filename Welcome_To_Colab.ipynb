```
import re

def stem_sentence(roots, sentence):
    # Create a mapping of root words
    root_mapping = {root: root for root in roots}

    # Create a regex pattern to match the roots (case-insensitive)
    pattern = re.compile(r'\b(' + '|'.join(root_mapping.keys()) + r')\b', re.IGNORECASE)

    # Function to replace words in the sentence with their roots
    def replace_with_roots(match):
        # Get the matched word
        word = match.group(0)
        # Return the corresponding root word
        return root_mapping[word.lower()]

    # Substitute the words in the sentence
    stemmed_sentence = pattern.sub(replace_with_roots, sentence)

    return stemmed_sentence

# Example usage
roots = ["cat", "rat", "bat"]
sentence = "the cattle was rattled by the battery"
output = stem_sentence(roots, sentence)

print("Original Sentence:", sentence)
print("Stemmed Sentence:", output)
```

```
import nltk
from nltk.stem import PorterStemmer

def stem_sentence(sentence, roots):
    """Stems a sentence using the Porter stemmer, ensuring root words from the given list are preserved.

    Args:
        sentence (str): The input sentence.
        roots (list): A list of root words to be preserved.

    Returns:
        str: The stemmed sentence.
    """

    stemmer = PorterStemmer()
    words = sentence.split()

    stemmed_words = []
    for word in words:
        stemmed_word = stemmer.stem(word)
        if stemmed_word in roots:
            stemmed_words.append(word)  # Preserve root words
        else:
            stemmed_words.append(stemmed_word)

    return " ".join(stemmed_words)

# Example usage
roots = ["cat", "rat", "bat"]
sentence = "the cattle was rattled by the battery"

stemmed_sentence = stem_sentence(sentence, roots)
print(stemmed_sentence)  # Output: "the cat was rat by the bat"
```
